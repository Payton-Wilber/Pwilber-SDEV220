```python
#Payton Wilber
#GPA Evaluator.py
# This app asks the user to input the first and last name of a student along with their GPA. The app then compares the student's GPA to the GPA requirements of the Honor Roll and Dean's list. Finally the app outputs whether opr not the student qualifies for thwe Honor Roll or Dean's list
print("Enter 'ZZZ' for the last name to quit.")

while True:
    last_name = input("\nEnter student's last name: ").strip().upper()
    
    if last_name == 'ZZZ':
        print("Goodbye!")
        break

    first_name = input("Enter student's first name: ").strip().capitalize()

    try:
        gpa_input = input("Enter student's GPA (e.g., 3.85): ")
        gpa = float(gpa_input)
    except ValueError:
        print("Invalid GPA entered. Please enter a number (e.g., 3.85).")
        continue 

    if gpa >= 3.5:
        print(f"\n*** Congratulations, {first_name} {last_name}! You have made the DEAN'S LIST! ***")
    

    elif gpa >= 3.25:
        print(f"\n*** Congratulations, {first_name} {last_name}! You have made the HONOR ROLL! ***")
    else:
        print(f"\n{first_name} {last_name} does not qualify for the Dean's List or Honor Roll this semester.")
```
