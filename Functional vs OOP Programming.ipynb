```python
#Sort 0s, 1s and 2s
#Could you come up with a one-pass algorithm using only constant extra space?: Yes
class Solution:
    def sort012(self, arr):
        low = 0    
        mid = 0    
        high = len(arr) - 1 

        while mid <= high:
            current_element = arr[mid]

            if current_element == 0:
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1
            elif current_element == 1:
                mid += 1
            elif current_element == 2:
                arr[high], arr[mid] = arr[mid], arr[high]
                high -= 1
    
        return arr 
```
